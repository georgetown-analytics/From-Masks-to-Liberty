# (3) 2020 US Presidential Election Data - County level votes
Wrangling the data set to merge to the other files

```python
#https://www.kaggle.com/unanimad/us-election-2020?select=president_county_candidate.csv
```

```python
president = pd.read_csv('s3://covidstatedata/president_county_candidate.csv')
```

```python
president.sort_values(by='state')
```

```python
#Parse out the election data by candidates of interest Joe Biden (Democrat) and Donald Trump (Republican); 
joe = president[president['candidate']=='Joe Biden']
don = president[president['candidate']=='Donald Trump']
```

```python
joe.sort_values(by=['state', 'county'])
```

```python
don.sort_values(by=['state','county'])
```

# Feature Engineering our target  from the election data 

```python
merged = pd.merge(joe, don, on=['state', 'county'], how='outer')
merged.sort_values(by=['state','county'])
```

```python
#calculating the percent of the counties votes were for Joe Biden v. Donald Trump
merged['% Joe'] = merged['total_votes_x']/(merged['total_votes_x']+merged['total_votes_y'])
```

```python
joes = merged[merged['% Joe'] >= .5]
dons = merged[merged['% Joe'] <= .5]
```

```python
joes.sort_values(by=['state', 'county'])
```

```python
dons.sort_values(by=['state','county'])
```

```python
joes['Dem0_Rep1'] = '0'
```

```python
dons['Dem0_Rep1'] = '1'
```

```python
dons=dons.sort_values(by=['state','county'])
dons
```

```python
#joes = joes.reset_index()
#dons = dons.reset_index()
```

```python
joes.drop('candidate_y', axis=1, inplace=True)
joes.drop('party_y', axis=1, inplace=True)
joes.drop('won_y', axis=1, inplace=True)
joes.drop('total_votes_y', axis=1, inplace=True)
```

```python
joes = joes.rename(columns={"candidate_x": "candidate", "party_x": "party", "total_votes_x": "total_votes", "won_x": "won"})
```

```python
dons.drop('candidate_x', axis=1, inplace=True)
dons.drop('party_x', axis=1, inplace=True)
dons.drop('total_votes_x', axis=1, inplace=True)
dons.drop('won_x', axis=1, inplace=True)
```

```python
dons = dons.rename(columns={"candidate_y": "candidate", "party_y": "party", "total_votes_y": "total_votes", "won_y": "won"})
```
